# แปลง NAN ของ Floor โดยใช้ KNNImputer

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
# numeric_col = [col for col in df.columns if pd.api.types.is_numeric_dtype(df[col])]

numeric_col = [col for col in df.columns if pd.api.types.is_numeric_dtype(df[col])]
```

```python
numeric_col
```

```python
imputer.fit(df[numeric_col])
```

```python
df[numeric_col] = imputer.transform(df[numeric_col])
```
